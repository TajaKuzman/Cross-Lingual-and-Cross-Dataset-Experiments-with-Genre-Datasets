```python
import pandas as pd
import json

with open("G:/My Drive/Institut Jozef Stefan/Projekti na Institutu/Določanje žanrov/Mapping to CORE/GINCO-1.0-suitable.json") as file:
    dataset = json.load(file)
```

```python
label_distribution = dataset_df.GINCORE.value_counts().to_dict()
for key in label_distribution:
    print(label_distribution[key])
```

```python
GINCORE_mapping = {'News/Reporting':'News', 'Promotion of Services':'Promotion', 'Opinionated News':'News', 'Promotion of a Product':'Promotion', 'Invitation':'Promotion'}
```

```python
for i in dataset:
    if i["primary_level_1"] == "Opinionated News" or i["primary_level_1"] == "News/Reporting":
        i["GINCORE"] = "News"
    elif i["primary_level_1"] == 'Promotion of Services' or i["primary_level_1"] == 'Promotion of a Product' or i["primary_level_1"] == 'Invitation':
        i["GINCORE"] = 'Promotion'
    else:
        current_label = i["primary_level_1"]
        i["GINCORE"] = current_label
```

```python
new_GINCO_file = open("GINCO-mapped-to-GINCORE.json", "w")
json.dump(dataset,new_GINCO_file, indent= "")
new_GINCO_file.close()
```
